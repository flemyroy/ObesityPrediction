# Data Cleaning

```python
roundattributes=['FCVC','NCP','CH2O','FAF','TUE']
for i in roundattributes:
    print(df[i].value_counts())
```

```python
roundattributes=['FCVC','NCP','CH2O','FAF','TUE']
for i in roundattributes:
    df[i]=round(df[i])
```

```python
df['BMI']=df['Weight']/(df['Height']*df['Height'])
df['BMI']=round(df['BMI'])
```

```python
df=df.drop(['Weight','Height'],axis=1)
```

```python
sns.boxplot(df['BMI'])
```

```python
Q1 = df.Age.quantile(0.25)
Q3 = df.Age.quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
df2 = df[(df['Age']<lower) | (df['Age']>upper)]
print("Number of Outliers Detected in Age:",df2.shape[0])
```

```python
for i,data in df.iterrows():
    if data['Age']<Q1:
        df.at[i,'Age']=lower  
    elif data['Age']>Q3:
        df.at[i,'Age']=upper
```

```python
#After removal of outliers
sns.boxplot(df['Age'])
```

```python
for i in df.columns:
    df[i]=df[i].astype('category')
```

```python
df.info()
```

```python
#Discretization
for i in df.columns:
    df[i]=df[i].cat.codes
    
df.head()
```

# ML Algorithms

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection  import train_test_split
```

```python
df.head()
```

```python
X=df.drop("NObeyesdad",axis=1)
y=df["NObeyesdad"]
```

```python
#Splitting into training and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Standardising the Attributes
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Decision Tree

```python
#Using Standard scaled data
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
model=DT.fit(X_train,y_train)
text_visual=tree.export_text(model)
fig=plt.figure(figsize=(25,20))
tree.plot_tree(model,filled=True)
fig.savefig("C:\\Users\\FLEMY\\Desktop\\Xaviers\\SemII\\Enabling Technologies (Practical)\\Obesity_Enc\\DT.png")
```

```python
y_pred_DT=DT.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
cm = confusion_matrix(y_test, y_pred_DT)
print(cm)
```

```python
acc_score_DT=accuracy_score(y_test,y_pred_DT)
acc_score_DT=round(acc_score_DT*100,2)
print(acc_score_DT)
```

```python
print(classification_report(y_test, y_pred_DT))
```

# Random Forest Classifier

```python
#Using Standard Scaled Data
model=RandomForestClassifier(n_estimators=500,max_depth=5)
model.fit(X_train,y_train)
```

```python
y_pred_RF=model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_RF)
print(cm)
```

```python
acc_score_RFC=accuracy_score(y_test,y_pred_RF)
acc_score_RFC=round(acc_score_RFC*100,2)
```

```python
print(classification_report(y_test, y_pred_RF))
```
